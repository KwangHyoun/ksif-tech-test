```python
#import libraries

import sys
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#reading csv file

df = pd.read_csv('/Users/gangmuk/remote_repo/ksif-tech-test/price.csv')
#print (df.describe)
#print (df.head())
#print(df)
```

```python
# Creating list including valid price values only

#print(df.iloc[:,0])
prices = []
prices = df.iloc[:,1]
#print (prices)
prices = [i for i in prices if pd.isnull(i) == False]
prices.reverse()
print (prices)
```

```python
# test example
# prices = [500.0, 750.0, 400.0, 600.0, 350.0, 800.0]

prev = prices[0]
drawdown_gap = 0.0
highest = 0.0
for cur in prices:
    highest = max(highest, cur)
    gap = highest - cur
    if gap > drawdown_gap:
        drawdown_gap = gap
        
print("mdd:", round(-drawdown_gap/highest*100.0, 2),"%")
```
